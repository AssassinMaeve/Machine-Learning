```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
```

```python
survive = titanic.groupby(["class",'sex'])["survived"].mean().unstack()
```

```python
survive.plot(kind='bar', stacked= True, color =['lightpink','red'])
```

```python
sns.kdeplot(data=titanic,x='age', hue = 'survived', fill=True)
```

```python
sns.violinplot(data=titanic, x='age', y='class')
```

```python
sns.swarmplot(data=titanic, x= 'class', y = 'age')
```

```python
numeric = titanic.select_dtypes(include=['float64','int64'])
```

```python
sns.heatmap(numeric.corr(), annot=True)
```
